## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
________________________________________________________________________________
A Primary Key uniquely identifies each record in a table, ensuring that no two rows have the same key value. It enforces entity integrity by preventing null values in its column. A Foreign Key, on the other hand, is a field (or collection of fields) in one table that refers to the Primary Key in another table, establishing a relationship between the two tables. This linkage ensures referential integrity, meaning that a Foreign Key value must either match an existing Primary Key value or be null

## Question 3

Explain the different types of relationships entities can have in a SQL database.

Type your answer below this line:
________________________________________________________________________________
One-to-One (1:1): Each entity in Table A is related to one entity in Table B and vice versa.
One-to-Many (1:N): An entity in Table A can relate to multiple entities in Table B, but each entity in Table B relates to only one entity in Table A.
Many-to-Many (M:N): Entities in Table A can relate to multiple entities in Table B and vice versa, typically managed through an associative entity or join table

## Question 4

Explain the various types of JOINs possible with SQL.

Type your answer below this line:
________________________________________________________________________________
INNER JOIN: Returns records with matching values in both tables.
LEFT JOIN (or LEFT OUTER JOIN): Returns all records from the left table and matched records from the right table; unmatched records from the right will show as null.
RIGHT JOIN (or RIGHT OUTER JOIN): Returns all records from the right table and matched records from the left; unmatched records from the left will show as null.
FULL JOIN (or FULL OUTER JOIN): Returns all records when there is a match in either left or right table records; unmatched records will show as null.
CROSS JOIN: Returns the Cartesian product of both tables, combining every row of one table with every row of another

## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
________________________________________________________________________________
Aggregate functions perform calculations on a set of values and return a single value, such as SUM(), AVG(), COUNT(), etc. The GROUP BY statement is used to arrange identical data into groups, allowing aggregate functions to be applied to each group separately. For instance, if you want to calculate the average salary per department, you would group by the department column and then apply the AVG() function to get results for each department

## Question 6

What role do Associative Entities play (JOIN Tables) in many-to-many JOINs?


Type your answer below this line:
________________________________________________________________________________
Associative entities, also known as join tables, serve as an intermediary between two entities that share a many-to-many relationship. They contain foreign keys referencing the primary keys of both related tables, enabling the establishment of connections between them. For instance, if students can enroll in multiple courses and courses can have multiple students, an associative entity like "Enrollments" would hold references to both "Students" and "Courses," effectively managing this complex relationship
